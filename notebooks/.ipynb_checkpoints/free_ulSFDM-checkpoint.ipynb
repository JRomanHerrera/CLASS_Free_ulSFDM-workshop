# Evolución del fondo cosmológico (`backgroun.c`)

Una vez definidos los parámetros del modelo, CLASS resuelve la evolución del fondo cosmológico integrando un sistema de ecuaciones diferenciales para las variables del campo escalar ultraligero.

En términos del número de e-folds, $N=\ln a$, el sistema está dado por


\begin{align}
y_1' &= \frac{3}{2}\left(1+\omega_{\rm tot}\right)y_1,\\
\alpha' &= \frac{3}{2}\left(\omega_{\rm tot}+\cos\theta\right),\\
\theta' &= -3\sin\theta+y_1,
\end{align}


donde la prima denota derivada respecto a $N$, $\omega_{\rm tot}$ es el parámetro de ecuación de estado efectivo del Universo y $(y_1,\theta,\alpha)$ son las variables dinámicas que describen la evolución del campo escalar.

En la implementación de CLASS, este sistema se encuentra en el archivo `background.c`, donde las derivadas de las variables se calculan mediante el siguiente bloque de código:

A continuación utilizaremos estas soluciones para estudiar la evolución temporal de las variables $y_1$ y $\theta$, para diferentes masas del campo escalar, permitiendo visualizar el inicio de las oscilaciones y la transición hacia el régimen de materia oscura efectiva.

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4), constrained_layout=True)

# ============================================================
# y1
# ============================================================
ax = axs[0]

for mass in masas:
    bg_sfdm = cosmos[mass].get_background()
    
    z_sfdm = bg_sfdm["z"]
    a_sfdm = 1.0 / (1.0 + z_sfdm)
    loga_sfdm = np.log10(a_sfdm)
    y1_sfdm = bg_sfdm["y1_sfdm"]

    ax.semilogy(loga_sfdm, y1_sfdm, color=model_colors[mass], lw=1.5, label=rf"$\mathcal{{M}}={mass}$")

ax.set_xlim(-14, 0)
ax.set_xlabel(r"$\log(a)$")
ax.set_ylabel(r"$y_1=2m_\phi/H$")
ax.set_title(r"Evolución de parámetro auxiliar $y_1$")
ax.grid(True, alpha=0.3)
ax.legend(fontsize=8, ncol=1, loc='upper left')

# ============================================================
# theta
# ============================================================
ax = axs[1]

ax.axhline(y=np.pi/2, color='k', ls='--', lw=1, label=r"$\theta=\pi/2$")

for mass in masas:
    bg_sfdm = cosmos[mass].get_background()

    z_sfdm = bg_sfdm["z"]
    a_sfdm = 1.0 / (1.0 + z_sfdm)
    loga_sfdm = np.log10(a_sfdm)
    theta_sfdm = bg_sfdm["theta_sfdm"]

    ax.semilogy( loga_sfdm, theta_sfdm, color=model_colors[mass], lw=1.5,label=rf"$\mathcal{{M}}={mass}$")

ax.set_xlim(-14, 0)
ax.set_xlabel(r"$\log(a)$")
ax.set_ylabel(r"$\theta_\phi$")
ax.set_title(r"Evolución de parámetro auxiliar$\theta_\phi$")
ax.grid(True, alpha=0.3)
ax.legend(fontsize=8, ncol=1, loc='upper left')

plt.show()
```

## Densidad, presión y ecuación de estado del campo escalar

A partir de las variables dinámicas del modelo, la ecuación de estado del campo escalar ultraligero puede escribirse como

\begin{equation}
\omega_\phi
\equiv
\frac{p_\phi}{\rho_\phi}
=
-\cos\theta.
\end{equation}

Por lo tanto, el comportamiento efectivo del campo depende directamente de la variable angular $\theta$. Durante el régimen oscilatorio, $\omega_\phi\$ oscila rápidamente y su promedio temporal tiende al comportamiento de materia oscura fría.

En la implementación dentro de `CLASS`, la variable $\alpha$ se utiliza para definir la fracción de densidad del campo escalar mediante

\begin{equation}
\Omega_\phi=e^{2\alpha}.
\end{equation}

Como la densidad total antes de añadir la contribución del campo escalar está dada por

\begin{equation}
\rho_{\rm tot}^{(\rm sin\,\phi)}
=
\sum_j \rho_j,
\end{equation}

la densidad del campo se reconstruye a partir de $\Omega_\phi$ como

\begin{equation}
\rho_\phi
=
\frac{\Omega_\phi}{1-\Omega_\phi}
\sum_j \rho_j
=
\frac{e^{2\alpha}}{1-e^{2\alpha}}
\rho_{\rm tot}^{(\rm sin\,\phi)}.
\end{equation}

La presión se obtiene directamente de la ecuación de estado:

\begin{equation}
p_\phi
=
\omega_\phi\rho_\phi
=
-\cos\theta\,\rho_\phi.
\end{equation}

Estas expresiones se implementan en el archivo `background.c`. Es importante que la contribución de ulSFDM se calcule al final, porque la variable `rho_tot` utilizada en la expresión anterior debe contener primero la suma de las demás componentes cosmológicas.

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4), constrained_layout=True)
# ============================================================
# Ecuación de estado
# ============================================================
ax = axs[0]
# Referencia ΛCDM
ax.plot(loga,  p_cdm / rho_cdm, color=colors["cdm"], lw=2, label=r"${\rm CDM}$")

for mass in masas:
    bg_sfdm = cosmos[mass].get_background()
    z_sfdm = bg_sfdm["z"]
    a_sfdm = 1.0 / (1.0 + z_sfdm)
    loga_sfdm = np.log10(a_sfdm)
    rho_sfdm = bg_sfdm["(.)rho_sfdm"]
    p_sfdm   = bg_sfdm["(.)p_sfdm"]
    ax.plot(loga_sfdm, p_sfdm / rho_sfdm, color=model_colors[mass], ls='--', lw=1, label=rf"$\mathcal{{M}}={mass}$")

ax.set_xlim(-14, 0)
ax.set_ylim(-1.02, 1.02)

ax.set_xlabel(r"$\log(a)$")
ax.set_ylabel(r"$\omega=p/\rho$")
ax.set_title("Equación de Estado: CDM vs. Free-ulSFDM")
ax.grid(True, alpha=0.3)
ax.legend(fontsize=8, ncol=1, loc='upper right')

# ============================================================
# Densidad de energía
# ============================================================
ax = axs[1]
# Referencia ΛCDM
ax.semilogy(loga, rho_cdm, color=colors["cdm"], lw=2, label=r"$\rho_{\rm CDM}$")
ax.semilogy(loga, rho_lambda, color=colors["lambda"], lw=2, label=r"$\rho_\Lambda$")

for mass in masas:
    bg_sfdm = cosmos[mass].get_background()
    z_sfdm = bg_sfdm["z"]
    a_sfdm = 1.0 / (1.0 + z_sfdm)
    loga_sfdm = np.log10(a_sfdm)
    rho_sfdm = bg_sfdm["(.)rho_sfdm"]
    ax.semilogy(loga_sfdm, rho_sfdm, color=model_colors[mass], ls='--', lw=1.5, label=rf"$\mathcal{{M}}={mass}$")

ax.set_xlim(-14, 0)
ax.set_ylim(1e-8, 1e18)

ax.set_xlabel(r"$\log(a)$")
ax.set_ylabel(r"$(8\pi G\,\rho/3)\,[{\rm Mpc}^{-2}]$")
ax.set_title("Densidad de energía")
ax.grid(True, which="both", alpha=0.3)
ax.legend(fontsize=8, ncol=1, loc='upper right')
plt.show()

fig, axs = plt.subplots(1, 2, figsize=(10, 4), constrained_layout=True)
# ============================================================
# EDensidad de energía normalizada
# ============================================================
ax = axs[0]
# Referencia ΛCDM
ax.axhline(y=1, color='k', ls='-', lw=1)
for mass in masas:
    bg_sfdm = cosmos[mass].get_background()
    z_sfdm = bg_sfdm["z"]
    a_sfdm = 1.0 / (1.0 + z_sfdm)
    loga_sfdm = np.log10(a_sfdm)
    rho_sfdm = bg_sfdm["(.)rho_sfdm"]
    p_sfdm   = bg_sfdm["(.)p_sfdm"]
    ax.plot(loga_sfdm, (rho_sfdm / rho_sfdm[-1])*(a_sfdm)**3, color=model_colors[mass], ls='-', lw=1, label=rf"$\mathcal{{M}}={mass}$")

ax.set_xlim(-14, 0)
ax.set_ylim(0, 1.5)

ax.set_xlabel(r"$\log(a)$")
ax.set_ylabel(r"$(\rho_\phi/\rho_{\phi 0})a^3$")
ax.set_title("Densidad de energía normalizado con el valor actual")
ax.grid(True, alpha=0.3)
ax.legend(fontsize=8, ncol=1, loc='upper right')

# ============================================================
# Densidad de energía Omega
# ============================================================
ax = axs[1]
# Referencia ΛCDM
ax.plot(loga, Omega_cdm , color=colors["cdm"], lw=2, label=r"CDM")
for mass in masas:
    bg_sfdm = cosmos[mass].get_background()
    z_sfdm = bg_sfdm["z"]
    a_sfdm = 1.0 / (1.0 + z_sfdm)
    loga_sfdm = np.log10(a_sfdm)
    Omega_sfdm = bg_sfdm["Omega_sfdm"]
    ax.plot(loga_sfdm, Omega_sfdm, color=model_colors[mass], ls='--', lw=1.5, label=rf"$\mathcal{{M}}={mass}$")

ax.set_xlim(-8, 0)
ax.set_ylim(0, 1)

ax.set_xlabel(r"$\log(a)$")
ax.set_ylabel(r"$\Omega=\rho/\rho_{\rm crit}$")
ax.set_title("Densidad $\Omega$: CDM Vs. Free-ulSFDM")
ax.grid(True, alpha=0.3)
ax.legend(fontsize=8, ncol=1, loc='upper left')

plt.show()
```

# Perturbaciones lineales (`perturbations.c`)

Una vez determinada la evolución del fondo cosmológico, CLASS resuelve la evolución de las perturbaciones lineales de cada componente. Para el campo escalar ultraligero, la dinámica puede escribirse en términos de dos variables auxiliares, $\delta_0$ y $\delta_1$, que describen las perturbaciones de densidad y de fase del campo en la parametrización utilizada para ulSFDM.

En términos del número de e-folds, $N=\ln a$, el sistema está dado por:

\begin{equation}
\begin{pmatrix}
\delta_0\\[1mm]
\delta_1
\end{pmatrix}' =
\begin{pmatrix}
k_{\rm eff}^2\sin\theta & -\left[3\sin\theta +k_{\rm eff}^2(1-\cos\theta) \right] \\[3mm]
k_{\rm eff}^2(1+\cos\theta) & -\left[ 3\cos\theta + k_{\rm eff}^2\sin\theta \right]
\end{pmatrix}
\begin{pmatrix}
\delta_0\\[1mm]
\delta_1
\end{pmatrix}
- \frac{\bar{h}'}{2}
\begin{pmatrix}
1-\cos\theta\\[1mm]
\sin\theta
\end{pmatrix},
\end{equation}

donde $k^2_{\rm eff}=k^2/k^2_{\rm J}= k^2/a^2H^2y_1$, es un parámetro que compara la escala del modo de Fourier $k$ con la escala característica de Jeans del campo escalar.
Las variables auxiliares $\delta_0$ y $\delta_1$ se definen con mayor detalle en las Eqs.(16) de arxiv:2307.05600. A partir de ellas, CLASS reconstruye las perturbaciones físicas del campo escalar, como el contraste de densidad y la divergencia de la velocidad.

La dependencia con $k_{\rm eff}^2$ permite distinguir dos regímenes físicos:
- Para $k_{\rm eff}^2\ll 1$, los modos se encuentran por encima de la escala de Jeans y el campo puede agruparse de manera semejante a la materia oscura fría.
- Para $k_{\rm eff}^2\gtrsim 1$, los efectos de gradiente se vuelven relevantes y pueden producir oscilaciones o supresión en el crecimiento de las perturbaciones.

En la implementación de CLASS, este sistema se encuentra en el archivo `perturbations.c`, donde las derivadas de $\delta_0$ y $\delta_1$ se calculan numéricamente junto con las perturbaciones del resto de las componentes cosmológicas.

Una vez que CLASS resuelva numéricamente las variables auxiliares $\delta_0$ y $\delta_1$, es posible reconstruir las perturbaciones físicas del campo escalar. En particular, pueden obtenerse el contraste de densidad, el contraste de presión y la divergencia de la velocidad, los cuales constituyen las cantidades utilizadas para comparar el modelo con las observaciones cosmológicas.

Estas cantidades están dadas por:

\begin{align}
\delta_{\rho\phi} =& \frac{\dot{\phi}\delta\dot{\phi}-\partial_\phi V(\phi)\ \delta\phi}{\rho_\phi}= \delta_0, \\
\delta_{p\phi} =& \frac{\dot{\phi}\delta\dot{\phi}+\partial_\phi V(\phi)\ \delta\phi}{\rho_\phi} = -\cos\theta \, \delta_0
+ \sin\theta\,\delta_1,\\
 \Theta_\phi =& \frac{k^2}{a}\frac{\delta\phi}{\dot{\phi}}=\frac{k^2}{a Hy_1}\frac{\rho_\phi}{(p_\phi +\rho_\phi)}\left[ (1-\cos\theta)\delta_1 - \sin\theta\delta_0 \right] ,
\end{align}

La primera ecuación muestra que la variable auxiliar $\delta_0$ coincide directamente con el contraste de densidad del campo escalar. Por otro lado, el contraste de presión depende tanto de $\delta_0$ como de $\delta_1$.

Estas expresiones son las implementadas en el archivo `perturbations.c`, donde las variables auxiliares se transforman en las perturbaciones físicas que posteriormente se utilizan para calcular observables cosmológicos, como el espectro de potencia de materia y las anisotropías del fondo cósmico de microondas.

## Las anisotropías del Fondo Cósmico de Microondas

El espectro de anisotropías de temperatura del Fondo Cósmico de Microondas (CMB) se expresa habitualmente mediante la cantidad

$D_{\ell}^{TT}= \frac{\ell(\ell+1)}{2\pi}\,C_{\ell}^{TT},$

donde $C_{\ell}^{TT}$ es el espectro angular de potencia de las anisotropías de temperatura. Como CLASS devuelve $C_{\ell}^{TT}$ en unidades adimensionales, el resultado se convierte convencionalmente a $\mu{\rm K}^2$ multiplicando por el cuadrado de la temperatura media del CMB:

$ D_{\ell}^{TT} = \left(T_{\rm CMB}\times10^{6}\right)^2 \frac{\ell(\ell+1)}{2\pi} C_{\ell}^{TT},$

con

$T_{\rm CMB}=2.7255\,{\rm K}.$

La cantidad $D_{\ell}^{TT}$ representa aproximadamente la contribución de cada multipolo $\ell$ a la varianza de las anisotropías de temperatura.

```python
fig, ax = plt.subplots(figsize=(10, 6), dpi=200)

# --- LCDM ---
Tcmb = 2.7255
cls_lcdm = cosmo[LCDM].lensed_cl(2500) # Calcula los espectros angulares del CMB hasta l=2500.
ll_lcdm = cls_lcdm['ell'][2:]          # Multipoles l
clTT_lcdm = cls_lcdm['tt'][2:]         # Obtiene los coeficientes del espectro angular de temperatura C_l^TT
Dl_lcdm = (Tcmb * 1e6)**2 * clTT_lcdm * ll_lcdm*(ll_lcdm+1)/(2*np.pi)
# Grafica
ax.plot(ll_lcdm, Dl_lcdm, color=model_colors["LCDM"], linewidth=2, label=r'$\Lambda$CDM')

# --- Modelos ulSFDM ---
for mass in masas:
    cls = cosmos[mass].lensed_cl(2500)
    ll = cls['ell'][2:]
    clTT = cls['tt'][2:]
    Dlm = (Tcmb*1e6)**2 * clTT * ll*(ll+1)/(2*np.pi)
    ax.plot(ll,Dlm, color=model_colors[mass], linestyle="--", linewidth=1, label=rf"$\mathcal{{M}}={mass}$")

# --- Formato de la Gráfica ---
ax.set_xscale('log')
ax.set_xlabel(r'$\text{Multipole } \ell$') # Añadido para identificar el eje X logarítmico
ax.set_ylabel(r'$D_\ell^{TT}\,[\mu\mathrm{K}^2]$')

ax.set_xlim(2, 2.5e3)
ax.legend(loc="upper left", frameon=True, framealpha=0.9) # Corregido loc numérico por string seguro

plt.tight_layout()
# fig.savefig('CMB_Anisotropies.pdf', dpi=300, bbox_inches='tight') 
plt.show()
```

#### Aunque la naturaleza de la materia oscura sea distinta, todos estos modelos producen prácticamente el mismo patrón de anisotropías del CMB.
